## Country vs Happiness (Before 2006)

```python
import matplotlib.pyplot as plt

# Filter data before 2006
data_2005 = df[df["year"] < 2006]

# Take average happiness per country
avg_data = data_2005.groupby("Country name")["Life Ladder"].mean()

# Plot
plt.figure(figsize=(10,6))
avg_data.plot(kind="bar", color="skyblue")

plt.title("Country vs Happiness (Before 2006)")
plt.xlabel("Country")
plt.ylabel("Life Ladder")

plt.xticks(rotation=90)
plt.show()
```

## INSIGHTS

#### This chart shows the happiness level of India for each year
#### We can see how happiness changes year by year
#### Some years show increase, while some show decrease

## yearly change in happiness

```python
year_avg = df.groupby("year")["Life Ladder"].mean()

change = year_avg.diff()

plt.figure()
plt.plot(change.index,change.values)
plt.title("Yearly change in happiness")
plt.xlabel("year")
plt.ylabel("change in Life Ladder")
plt.axhline(0)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter India data
india = df[df['Country name'] == 'India']

# Sort by year (important)
india = india.sort_values('year')

# Plot
plt.figure(figsize=(8,5))
plt.bar(india['ear'], india['Life Ladder'], color='skyblue')
plt.title('year-wise Happiness Level in India')
plt.xlabel('year')
plt.ylabel('Happiness Score')

plt.xticks(rotation=45)
plt.show()
```

## INSIGHTS

```python
import matplotlib.pyplot as plt

# Select few countries
countries = ['India', 'Finland', 'United States', 'Brazil']

data = df[df['Country name'].isin(countries)]

# Average happiness per country
avg_data = data.groupby('Country name')['Life Ladder'].mean()

# Plot
plt.figure(figsize=(7,5))
avg_data.plot(kind='bar', color=['blue','green','orange','red'])

plt.title('Average Happiness Comparison')
plt.xlabel('Country')
plt.ylabel('Life Ladder (Happiness Score)')

plt.xticks(rotation=0)
plt.show()
```

## INSIGHTS

#### This chart compares happiness scores between different countries

#### Finland has higher happiness, while India has moderate levels

#### This shows differences in economic and social conditions
